```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

page = requests.get('https://en.wikipedia.org/wiki/List_of_association_footballers_who_died_while_playing').text
soup = BeautifulSoup(page, 'html.parser')
table = soup.find('table', class_="wikitable sortable")

df = pd.read_html(str(table))
df = pd.concat(df)

df_year_deaths = df['Date'].apply(lambda x: int(x[-4:])).value_counts().reset_index()\
    .rename(columns={'index': 'Year', 'Date': 'Sudden Deaths in Football'})

df_year_deaths.sort_values('Year').plot(kind = 'scatter', x = 'Year', y = 'Sudden Deaths in Football');
```
